## Perform data cleaning and preprocessing, including handling missing values and encoding categorical variables.

### Encoding categorical variables.

```python
data.describe()
```

#### so our cetrgorical columns are ['Garage Area', 'YearBuilt']

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create LabelEncoder object
encoder = LabelEncoder()

# Encode categorical columns
categorical_features = ['GarageArea', 'YearBuilt']
encoded_data = data[categorical_features].apply(encoder.fit_transform)

# Add encoded columns to the original DataFrame
data[categorical_features] = encoded_data
```

```python
data.head()
```

```python
# Drop 'Id' column as it's not required

data = data.drop('ID', axis=1)
```

### Other Optional steps: to improve performance

1. Scaling
2. feature selection

```python
X=data.drop('SalePrice', axis=1)
Y=data['SalePrice']
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit and Transform the data using the scaler
scaled_data = scaler.fit_transform(X)

# Replace the original data with the scaled data
X_scaled=X.copy()
X_scaled.update(pd.DataFrame(scaled_data, columns=X.columns))
```

```python
X_scaled.head()
```

#### above looks good, now we can select k best features

```python
from sklearn.feature_selection import SelectKBest, f_regression

# Create a SelectKBest object
selector = SelectKBest(score_func=f_regression, k=5)

# Fit and transform the data using the selector
X_selected = selector.fit_transform(X_scaled, Y)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_feature_indices]

print("Selected Features:", selected_features)
```

```python
X_final=X_scaled[selected_features]
X_final.head()
```

### Split the data into train and test sets


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, Y, test_size=0.2, random_state=42)
```

## **Task 2: Regression**

### Choose an appropriate regression algorithm and justify your choice,Train the model on the training data,Evaluate the model's performance using relevant metrics.



##### i haven't done much EDA, so will try some random possible regression algo to find out what is best performing

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# a function to train and evaluate a model
def train_and_evaluate(model_name, model):
  # Train the model
  model.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred = model.predict(X_test)

  # Calculate mean squared error (MSE)
  mse = mean_squared_error(y_test, y_pred)

  # Print results
  print(f"{model_name} MSE: {mse}")
```

```python
# Create a dictionary of models to try
models = {
  "Linear Regression": LinearRegression(),
  "Ridge Regression": Ridge(alpha=1),  # Adjust alpha for regularization
  "Lasso Regression": Lasso(alpha=0.1),  # Adjust alpha for regularization
  "Random Forest": RandomForestRegressor(n_estimators=100),  # Adjust hyperparameters
  "Support Vector Regression": SVR()  # Consider kernel and hyperparameter tuning
}

# Train and evaluate each model
for name, model in models.items():
  train_and_evaluate(name, model)
```
